```python
interested_columns = [
    	"acousticness","danceability","duration_ms","energy","instrumentalness","loudness","tempo",
        "speechiness","tempo","valence"
]
```

```python
for columns in interested_columns:
    pos_data = data[data['target'] == 1][columns]
    neg_data = data[data['target'] == 0][columns]

    plt.figure(figsize=(10, 8))

    sns.distplot(pos_data , bins = 40 , label='Positive' , color='Blue')
    sns.distplot(neg_data , bins = 40 , label='Negative' , color = 'green')

    plt.legend(loc='upper right')
    plt.title(f"Positive and Negative Histogram for {columns}")
    plt.show()
```
